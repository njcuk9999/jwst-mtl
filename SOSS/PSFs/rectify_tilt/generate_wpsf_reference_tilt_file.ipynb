# Run

```python
path = "/genesis/jwst/jwst-ref-soss/monochromatic_PSFs/psf_20200602/"
files_endwith = '.fits'
flist = []
for file in listdir(path):
    if file.endswith(files_endwith):
        flist.append(file)
print(len(flist))

os = 10
view = 'dms'

wavs = np.zeros(len(flist))
tilts = np.zeros(len(flist))
for i,file in enumerate(flist):
    
    os_from_file, psf_dim, wav = rot_wpsf.get_webbpsf_pars_from_filename(file)
    
    psf = rot_wpsf.get_webbpsf( wav*1e-6 , view=view , oversamp=os
                              , doPrint=True , doPlot=False , save=False, savepath=None
                              , psf_dim = psf_dim , jitter='gaussian' , jitter_sigma=0.007
                              , return_fits_HDUL=False
                              )
    tilt = rot_wpsf.get_webbpsf_tilt( psf , oversamp=os , view=view , fitrad=os
                                    , mode='ccf' , spatbox_size=23 , specbox_size=20
                                    , doPlot=False)
    wavs[i] = wav
    tilts[i] = tilt
wavs , inds = np.unique( wavs , return_index=True)
tilts = tilts[inds]
result = np.transpose(np.array([wavs,tilts]))
print(result.shape)
np.savetxt("webbpsf_tilt_os"+str(os)+"_nval"+str(len(result))+".txt" , result)
```
